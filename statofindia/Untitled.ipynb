```python
import requests
from bs4 import BeautifulSoup

# Define the URL
url = 'https://statisticstimes.com/'

try:
    # Send an HTTP GET request to the URL
    response = requests.get(url)
    response.raise_for_status()  # Check for any HTTP request errors

    # Create a BeautifulSoup object to parse the page
    soup = BeautifulSoup(response.text, 'html.parser')

    # Find the link to the "Economy" page
    economy_link = 'economy/india/indian-states-gdp.php'

    if economy_link:
        economy_url = 'http://statisticstimes.com/' + economy_link
        response = requests.get(economy_url)

        # Check for any HTTP request errors
        response.raise_for_status()

        # Create another BeautifulSoup object for the "Economy" page
        economy_soup = BeautifulSoup(response.text, 'html.parser')

        # Find the elements containing GSDP details
        table = economy_soup.find('table', {'id': 'table_id'})
        rows = table.find_all('tr')[2:]

        # Initialize lists to store the scraped data
        rank_list = []
        state_list = []
        gdp_1819_list = []
        gdp_1920_list = []
        share_1819_list = []
        gdp_billion_list = []

        # Iterate through the rows and extract the details
        for row in rows:
            columns = row.find_all('td')
            rank = columns[0].text.strip()
            state = columns[1].text.strip()
            gdp_1819 = columns[2].text.strip()
            gdp_1920 = columns[3].text.strip()
            share_1819 = columns[4].text.strip()
            gdp_billion = columns[5].text.strip()

            rank_list.append(rank)
            state_list.append(state)
            gdp_1819_list.append(gdp_1819)
            gdp_1920_list.append(gdp_1920)
            share_1819_list.append(share_1819)
            gdp_billion_list.append(gdp_billion)

        # Print or process the scraped data as needed
        for i in range(len(rank_list)):
            print(f"A) Rank: {rank_list[i]}")
            print(f"B) State: {state_list[i]}")
            print(f"C) GSDP(18-19): {gdp_1819_list[i]}")
            print(f"D) GSDP(19-20): {gdp_1920_list[i]}")
            print(f"E) Share(18-19): {share_1819_list[i]}")
            print(f"F) GDP($ billion): {gdp_billion_list[i]}")
            print()

    else:
        print("Economy link not found on the homepage.")

except requests.exceptions.RequestException as e:
    print(f"HTTP request error: {e}")

except Exception as e:
    print(f"An error occurred: {e}")
```
